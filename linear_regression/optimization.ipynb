```python
import matplotlib.pyplot as plt
import numpy as np

# example of an function 
def f(x):
    return x**2 + x + 1

# make data
x_1 = np.linspace(start=-3, stop=3, num=100)
plt.plot(x_1, f(x_1))
```

## Slope and derivatives


```python
def df(x):
    return 2*x + 1

plt.subplot(1, 2, 1)

x_1 = np.linspace(start=-3, stop=3, num=100)
plt.plot(x_1, f(x_1))
plt.subplot(1, 2, 2)
plt.grid()
plt.plot(x_1, df(x_1))
```

## find the lowest cost algorithm
Convex ( it has the lowest opposite to concave)

```python
# Gradient descent 
new_x = 3
previous_x = 0
step_multiplier = 0.1
precision = 0.00001
x_list = []
slope_list = []

for n in range(500):
    previous_x = new_x # where new_x is a random guess 
    gradient = df(previous_x)

    x_list.append(new_x)
    slope_list.append(df(new_x))

    # adjust
    new_x = previous_x - step_multiplier * gradient
    if abs(new_x - previous_x) < precision:
        print("ah oh")
        break

print("local minimum occurs at ", new_x)
print("cost is ", f(new_x))

x_1 = np.linspace(start=-3, stop=3, num=100)
plt.plot(x_1, f(x_1))
plt.scatter(x_list, f(np.array(x_list)))
```

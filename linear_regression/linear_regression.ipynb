````python
import pandas
from pandas import DataFrame
import matplotlib.pyplot
data = pandas.read_csv('cost_revenue_clean.csv')
data.describe()
````

````python
X = DataFrame(data, columns=['production_budget_usd'])
y = DataFrame(data, columns=['worldwide_gross_usd'])
import matplotlib.pyplot as plt


plt.scatter(X, y)
plt.title("example")
plt.xlabel('x')
plt.ylabel('y')
plt.show()
````

# theory of linear regression 
- In machine learning, those are Parameters, we are just giving estimates 

`h(x) = theta0 + theta1* x`
- `residual` could help us know how good the measure is 

the goal to minimize: 
```
RSS = (y1 - h(x1))^2 +  (y2 - h(x2))^2
```

# Use library



````python
# use sklearn
from sklearn.linear_model import LinearRegression


regression = LinearRegression()
regression.fit(X, y)
theta1 = regression.coef_  # theta_1

theta0 = regression.intercept_ # theta_0
````

````python
X = DataFrame(data, columns=['production_budget_usd'])
y = DataFrame(data, columns=['worldwide_gross_usd'])
import matplotlib.pyplot as plt

plt.plot(X, regression.predict(X), color='red')

plt.scatter(X, y)
plt.title("example")
plt.xlabel('budget')
plt.ylabel('revenue')
plt.show()
````

## How to evaluate the algorithm 
- why intercept is -7000000. which doesn't make sense, the prediction algorithm doesn't look good in extreame 
- we need `the goodness of fit`
    - means it can explains the % of the cases 

````python
regression.score(X, y) # it can explains 50% the data 
````
